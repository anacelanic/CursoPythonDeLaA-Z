La sucesión de Fibonacci es una sucesión infinita que se caracteriza porque cada término es la suma de los dos anteriores. Algunos de sus términos son 1, 1, 2, 3, 5, 8, 13...

Supongamos que queremos que se nos impriman los 20 primeros términos de esta serie. Por tanto, necesitaremos por un lado los términos de la serie y, por otro, los índices que ocupan.

**Ejercicio.** Pensad en cómo podríais resolver este problema en el cual se exige que en algún momento utilicéis el comando `break`.

```python
fibo_ant = 1 # Término anterior
fibo = 1 # Término actual
idx = 3 # Como ya tenemos los dos primeros términos, empezamos con el índice 3

print("El término {} ocupa la posición {}".format(fibo_ant, 1))
print("El término {} ocupa la posición {}".format(fibo, 2))

while fibo <= 500000 and idx <= 20: # Establecemos una cota para que el bucle no sea infinito
  temp = fibo  # Guardamos temporalmente el fibonacci actual 
  fibo = fibo + fibo_ant  # Calculamos el nuevo término de la sucesión
  fibo_ant = temp # Modicamos el valor del término anterior
  
  print("El término {} ocupa la posición {}".format(fibo, idx))
  
  idx += 1 # Incrementamos el valor del índice
```

y devuelva toda esa información en un diccionario.

```python
def contar_caracteres(string):
    caracteres = []
    apariciones = []
    for l in string:
        if l in caracteres:
            posicion = caracteres.index(l)
            apariciones[posicion] += 1
        else:
            caracteres.append(l)
            apariciones.append(1)
    return dict(zip(caracteres, apariciones))
```

##3
Dada una lista de palabras, quédate con filter() con las que tengan más vocales que consonantes. Necesitarás una función que devuelva si una palabra tiene más vocales que consonantes.

```python
def mas_vocales(w):
    vocales = ['a', 'e', 'i', 'o', 'u']
    contador = 0
    for i in w:
        if i in vocales:
            contador += 1
    consonantes =  len(w) - contador
    if consonantes < contador:
        return True
    return False

palabras = ['hola', 'queso', 'matematicas', 'ana', 'manzana', 'python', 'murcielago']

list(filter(mas_vocales, palabras))
```

una función que devuelva el número de divisores de un número dado

```python
enteros = [1, 2, -3, 4, 5, 6, -9, 30, 100]

def mas_5_divisores(entero):
    contador = 0
    for x in range(1, entero):
        if entero % x == 0:
            contador += 1
    if contador > 5:
        return True
    return False

list(filter(mas_5_divisores, enteros))
```

parámetros el día (day), mes (month) y año (year). Los 3 parámetros son de tipo int y por defecto todos

```python
import random
from datetime import date

total_dias_mes = {1: 31, 2: 28, 3:31, 4:30 , 5: 31, 6: 30, 7: 31, 8: 31, 9: 30, 10: 31, 11: 30, 12: 31}
month_in_english = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 
                            6: 'June', 7: 'July', 8: 'August', 9: 'September', 
                            10: 'October', 11: 'November', 12: 'December'}


def date_complete(entero, cifras):
    while cifras > len(str(entero)):
        entero = '0' + str(entero)
    return entero

#punto 1
class Date():

    """def __init__(self, day = 1, month = 1, year = 1):
        if type(day) is int and type(year) is int and type(month) is int:
            self.year = year
            self.month = month
            self.day = day
        else:
            print("Los parámetros {}, {} y {} no son enteros".format(day, month, year))"""

    #punto 2 junto con método date_complete
    def __str__(self):
        dia = self.day
        mes = self.month
        anio = self.year
        if mes < 10:
            mes = date_complete(self.month, 2)
        if dia < 10:
            dia = date_complete(self.day, 2)
        if anio < 1000:
            anio = date_complete(self.year, 4)
        return '{}/{}/{}'.format(dia, mes, anio)

    #punto 3
    def isLeap(self):
        print('{} {} año bisiesto'. format(self.year, 'es' if self.year % 4 == 0 else 'no es'))

    #punto 4
    def totalMonthDays(self):
       dias = total_dias_mes.get(self.month)
       print('El mes {} tiene {} dias'.format(self.month, '29' if self.month == 2 and self.year % 4 == 0 else dias))

 
    def validDate(self, day, month, year):
        if type(day) is int and type(year) is int and type(month) is int:
             if  month <= 12 and day <= total_dias_mes.get(month):
                return True
        return False
    
    def __init__(self, day = 1, month = 1, year = 1):
        if self.validDate(day, month, year):
            self.year = year
            self.month = month
            self.day = day
        else:
            print('¡¡¡La fecha introducida no es una fecha válida!!!')        


    #punto 5
    @property
    def monthName(self):
        return month_in_english.get(self.month)
        

    #punto 6
    @staticmethod
    def areEqual(date1, date2):
        if date1.day == date2.day and date1.month == date2.month and date1.year == date2.year:
            return True
        return False

    @staticmethod
    def isLater(date1, date2):
        if date1.year > date2.year:
            return True
        elif date1.year == date2.year:
            if date1.month > date2.month:
                return True
            elif date1.month == date2.month:
                if date1.day > date2.day:
                    return True
        return False
    
    @staticmethod
    def isPrevious(date1, date2):
        return Date.isLater(date2, date1)  


    #punto 7
    @classmethod
    def firstDayOfTheYear(cls, year):
        return Date(1, 1, year)

    @classmethod
    def lastDayOfTheYear(cls, year):
        return Date(31, 12, year)

    def plusDay(self):
        if Date.areEqual(Date.lastDayOfTheYear(self.year), self):
            self = Date.firstDayOfTheYear(self.year + 1)
        elif self.day == total_dias_mes.get(self.month):
            self.month = self.month + 1
            self.day = 1
        else:
            self.day += 1
        return self

    def minusDay(self):
        if Date.areEqual(Date.firstDayOfTheYear(self.year), self):
            self = Date.lastDayOfTheYear(self.year - 1)
        elif self.day == 1:
            self.month = self.month - 1
            self.day = total_dias_mes.get(self.month)
        else:
            self.day -= 1
        return self


    #punto 8
    @classmethod
    def copy(cls, date):
        return Date(date.day, date.month, date.year)

    @staticmethod
    def difference(d1, d2):
        if Date.isPrevious(d1, d2):
            anterior = date(d1.year, d1.month, d1.day)
            posterior = date(d2.year, d2.month, d2.day)
        else:
            posterior = date(d1.year, d1.month, d1.day)
            anterior = date(d2.year, d2.month, d2.day)
        return (posterior - anterior).days


    #punto 9
    @classmethod
    def randomDate(cls):
        year = random.randrange(0, 2100)
        month = random.randrange(1, 13)
        day = random.randrange(1, 30 if year % 4 == 0 and month == 2 else total_dias_mes.get(month) + 1)
        return cls(day, month, year)


    #punto 10
    @classmethod
    def toDate(cls, string):
        fecha = string.split()
        for key, value in month_in_english.items():
            if value == fecha[1]:
                mes = key
        return Date(int(fecha[0]), mes, int(fecha[2]))
```

```python
class Encoder():

    def __init__(self, key):
        self.key = key

    def encrypt(self, mensaje):
        return self.f(mensaje, self.key)
        
    def decrypt(self, mensaje):
        return self.f(mensaje, -self.key)

    def f(self, mensaje, key):
        msj = str(mensaje).split()
        contador = 0
        msj_encriptado = ''
        for palabra in msj:
            contador += 1
            for letra in palabra:
                caracter = ord(letra) + key
                msj_encriptado += chr(caracter)
            if contador != len(msj):
                msj_encriptado += ' '
        return msj_encriptado


        
```

```python
from numpy import random 

class Pirate():
    boat_life = 200

    def __init__(self, name, weapon):
        self.name = name
        self.weapon = weapon

    def attack(self, enemy):
        if  not (enemy.position == 'far' and self.weapon.scope == 'short'):
            print('{} ataca a {}'.format(self.name, enemy.name))
            enemy.hit(self.weapon.damage)
        else:
            print('El ataque no alcanza al enemigo')

    @classmethod
    def hit(cls, damage):
        Pirate.boat_life -= damage
        print('Vida del barco: {}'.format(Pirate.boat_life))


class Weapon():
    def __init__(self, name, damage, scope):
        self.name = name
        self.damage = damage
        if scope not in ['short', 'long']:
            raise ValueError('El alcance del arma solo puede tomar los valores short o long') 
        self.scope = scope


class Enemy():
    def __init__(self, name, health, position, min_damage, max_damage):
        self.name = name
        self.health = health
        if position not in ['near', 'far']:
            raise ValueError('El alcance de la posicion solo puede tomar los valores near o far') 
        self.position = position
        self.min_damage = min_damage
        self.max_damage = max_damage

    def hit(self, damage):
        self.health -= damage
        print('La vida de {} ahora es {}'.format(self.name, self.health))

    def attack(self, pirata):
        danio = random.randint(2)
        if danio == 0:
            print('{} ataca con el minimo daño a {}'.format(self.name, pirata.name))
            Pirate.hit(self.min_damage)
        else:
            print('{} ataca con el maximo daño a {}'.format(self.name, pirata.name))
            Pirate.hit(self.max_damage)
    
    def move(self):
        if self.position == 'near':
            print('{} cambio su posicion de near a far'.format(self.name))
            self.position = 'far'
        else:
            print('{} cambio su posicion de far a near'.format(self.name))
            self.position = 'near'            

#armas
sword = Weapon('Espada', 3, 'short')
ax = Weapon('Hacha', 5, 'short')
bow = Weapon('Arco', 2, 'long')

#enemys
torvellino1 = Enemy('Torvellino 1', 25, 'near', 2, 5)
torvellino2 = Enemy('Torvellino 2', 25, 'near', 4, 8)
torvellino3 = Enemy('Torvellino 3', 25, 'far', 3, 5)
torvellino4 = Enemy('Torvellino 4', 25, 'far', 6, 9)

#piratas
pyratilla = Pirate('Pyratilla', sword)
pym = Pirate('Pym', bow)
pyerce = Pirate('Pyerce', ax)

#Batalla
piratas = [pyratilla, pym, pyerce]
enemys = [torvellino1, torvellino2, torvellino3, torvellino4]

def restart_battle():

    global contador
    contador = 0
    global sin_enemigos
    sin_enemigos = False

    print('\n Que comience la batalla!!! \n')
    for enemy in enemys:
        enemy.health = 25

    Pirate.boat_life = 200
        

def chequear_enemigos():
    global contador
    if contador < 4:
        if enemys[contador].health > 0:
            return True
        else:
            if contador < 3:
                contador += 1
                print('El enemigo {} fue eliminado'.format(enemys[contador].name))
                return True
    return False
    
```

• los métodos de clase .sum() y .subtract() para que devuelvan objetos de la clase Vector3D.

```python
    def __str__(self):
        return super().__str__()[:-1] + ", {})".format(self.z)
    
    @property
    def module(self):
        radicando = super().module**2 + math.sqrt(self.z)
        return math.sqrt(radicando)
    
    def scalar_prod(self, parameter = 1):
        self.z *= parameter
        super().scalar_prod(parameter)
        return self.__str__()
    
    @classmethod
    def sum(cls, v1, v2):
        return Vector3D(v1.x + v2.x, v1.y + v2.y, v1.z + v2.z)
    
    @classmethod
    def substract(v1, v2):
        return Vector3D(v1.x - v2.x, v1.y - v2.y, v1.z - v2.z)
    
    @staticmethod
    def dot_product(v1, v2):
        return super().dot_product(v1, v2) + v1.z * v2.z
    
    @classmethod
    def distance(cls, v1, v2):
        radicando = substract(v1, v2)
        return radicando.module()
```

```python
def a_radianes(num_degree):
    if num_degree < 0:
        while num_degree < 0:
            num_degree += 360
    elif num_degree > 360:
        while num_degree > 360:
            num_degree -= 360
    return num_degree * math.pi / 180

print(a_radianes(270) == math.radians(270))

def a_grados(num_radianes):
    if num_radianes < 0:
        while num_radianes < 0:
            num_radianes += math.pi*2
    elif num_radianes > 2*math.pi:
        while num_radianes > 2*math.pi:
            num_radianes -= 2*math.pi
    return num_radianes * 180 / math.pi

print(a_grados(270) == math.degrees(270))

print(a_radianes(270))
print(a_radianes(630))
print(a_grados(2*math.pi))
print(a_grados(4*math.pi))
```

Dada la función divisors(), asegura con assert que el parámetro n se trata de un número entero positivo.

De lo contrario, muestra como mensaje que "n debe ser de tipo int y debe ser mayor que 0"


```python
def divisors(n):
    """
    Calcula los divisores de un número entero positivo.
    Args:
        n: Número entero positivo
    Returns:
        divisors: Lista de divisores de n
    """
    assert (type(n) == int and n > 0), 'n debe ser de tipo int y debe ser mayor que 0'
    divisors = []
    for i in range(1, n + 1):
        if n % i == 0: divisors.append(i)
    return divisors
```

##Ej 2
Dada la función divisors(), muestra TypeError con el mensaje correspondiente si el parámetro n no se
trata de un número entero y muestra ValueError con el mensaje correspondiente si n no se trata de un
número positivo.

```python
def divisors(n):
    """
    Calcula los divisores de un número entero positivo.
    Args:
        n: Número entero positivo
    Returns:
        divisors: Lista de divisores de n
    """
    if not type(n) == int:
        raise TypeError('n debe ser entero')
    if not n > 0:
        raise ValueError('n debe ser mayor que cero')
    divisors = []
    for i in range(1, n + 1):
        if n % i == 0: divisors.append(i)
    return divisors
```
