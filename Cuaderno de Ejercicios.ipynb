La sucesi√≥n de Fibonacci es una sucesi√≥n infinita que se caracteriza porque cada t√©rmino es la suma de los dos anteriores. Algunos de sus t√©rminos son 1, 1, 2, 3, 5, 8, 13...

Supongamos que queremos que se nos impriman los 20 primeros t√©rminos de esta serie. Por tanto, necesitaremos por un lado los t√©rminos de la serie y, por otro, los √≠ndices que ocupan.

**Ejercicio.** Pensad en c√≥mo podr√≠ais resolver este problema en el cual se exige que en alg√∫n momento utilic√©is el comando `break`.

```python
fibo_ant = 1 # T√©rmino anterior
fibo = 1 # T√©rmino actual
idx = 3 # Como ya tenemos los dos primeros t√©rminos, empezamos con el √≠ndice 3

print("El t√©rmino {} ocupa la posici√≥n {}".format(fibo_ant, 1))
print("El t√©rmino {} ocupa la posici√≥n {}".format(fibo, 2))

while fibo <= 500000 and idx <= 20: # Establecemos una cota para que el bucle no sea infinito
  temp = fibo  # Guardamos temporalmente el fibonacci actual 
  fibo = fibo + fibo_ant  # Calculamos el nuevo t√©rmino de la sucesi√≥n
  fibo_ant = temp # Modicamos el valor del t√©rmino anterior
  
  print("El t√©rmino {} ocupa la posici√≥n {}".format(fibo, idx))
  
  idx += 1 # Incrementamos el valor del √≠ndice
```

Con un bucle `while`, dados dos n√∫meros enteros proporcionados por el usuario, vamos a encontrar el primer n√∫mero divisible entre 2, 3 y 5, siempre que sea posible, que se encuentre dentro del intervalo formado por los dos n√∫meros dados por el usuario (ambos extremos tambi√©n incluidos).

```python
num1 = int(input("Ingreso un numero entero: "))
num2 = int(input("Ingrese un segundo numero entero: "))

if num1 >= num2:
    mayor = num1
    menor = num2
else: 
    mayor = num2
    menor = num1
i = menor
while i <= mayor:
    if i % 2 == 0 and i % 3 == 0 and i % 5 == 0:
        divisor  = i
        break

    i += 1
    
if i > mayor + 1:
    print("No existe un divisor de 2, 3 y 5 dentro del intervalor".format(menor, mayor))
else:
    print(divisor)
```

```python
from numpy import random 

class Pirate():
    boat_life = 200

    def __init__(self, name, weapon):
        self.name = name
        self.weapon = weapon

    def attack(self, enemy):
        if  not (enemy.position == 'far' and self.weapon.scope == 'short'):
            print('{} ataca a {}'.format(self.name, enemy.name))
            enemy.hit(self.weapon.damage)
        else:
            print('El ataque no alcanza al enemigo')

    @classmethod
    def hit(cls, damage):
        Pirate.boat_life -= damage
        print('Vida del barco: {}'.format(Pirate.boat_life))


class Weapon():
    def __init__(self, name, damage, scope):
        self.name = name
        self.damage = damage
        if scope not in ['short', 'long']:
            raise ValueError('El alcance del arma solo puede tomar los valores short o long') 
        self.scope = scope


class Enemy():
    def __init__(self, name, health, position, min_damage, max_damage):
        self.name = name
        self.health = health
        if position not in ['near', 'far']:
            raise ValueError('El alcance de la posicion solo puede tomar los valores near o far') 
        self.position = position
        self.min_damage = min_damage
        self.max_damage = max_damage

    def hit(self, damage):
        self.health -= damage
        print('La vida de {} ahora es {}'.format(self.name, self.health))

    def attack(self, pirata):
        danio = random.randint(2)
        if danio == 0:
            print('{} ataca con el minimo da√±o a {}'.format(self.name, pirata.name))
            Pirate.hit(self.min_damage)
        else:
            print('{} ataca con el maximo da√±o a {}'.format(self.name, pirata.name))
            Pirate.hit(self.max_damage)
    
    def move(self):
        if self.position == 'near':
            print('{} cambio su posicion de near a far'.format(self.name))
            self.position = 'far'
        else:
            print('{} cambio su posicion de far a near'.format(self.name))
            self.position = 'near'            

#armas
sword = Weapon('Espada', 3, 'short')
ax = Weapon('Hacha', 5, 'short')
bow = Weapon('Arco', 2, 'long')

#enemys
torvellino1 = Enemy('Torvellino 1', 25, 'near', 2, 5)
torvellino2 = Enemy('Torvellino 2', 25, 'near', 4, 8)
torvellino3 = Enemy('Torvellino 3', 25, 'far', 3, 5)
torvellino4 = Enemy('Torvellino 4', 25, 'far', 6, 9)

#piratas
pyratilla = Pirate('Pyratilla', sword)
pym = Pirate('Pym', bow)
pyerce = Pirate('Pyerce', ax)

#Batalla
piratas = [pyratilla, pym, pyerce]
enemys = [torvellino1, torvellino2, torvellino3, torvellino4]

def restart_battle():

    global contador
    contador = 0
    global sin_enemigos
    sin_enemigos = False

    print('\n Que comience la batalla!!! \n')
    for enemy in enemys:
        enemy.health = 25

    Pirate.boat_life = 200
        

def chequear_enemigos():
    global contador
    if contador < 4:
        if enemys[contador].health > 0:
            return True
        else:
            if contador < 3:
                contador += 1
                print('El enemigo {} fue eliminado'.format(enemys[contador].name))
                return True
    return False
    
```

‚Ä¢ el m√©todo de instancia .scalar_prod()

‚Ä¢ el m√©todo est√°tico .dot_product(), pues ahora el producto escalar se calcula como u ¬∑ v = uxvx +
uyvy + uzvz

‚Ä¢ el m√©todo de clase .distance(), pues ahora la distancia

```python
    def __str__(self):
        return super().__str__()[:-1] + ", {})".format(self.z)
    
    @property
    def module(self):
        radicando = super().module**2 + math.sqrt(self.z)
        return math.sqrt(radicando)
    
    def scalar_prod(self, parameter = 1):
        self.z *= parameter
        super().scalar_prod(parameter)
        return self.__str__()
    
    @classmethod
    def sum(cls, v1, v2):
        return Vector3D(v1.x + v2.x, v1.y + v2.y, v1.z + v2.z)
    
    @classmethod
    def substract(v1, v2):
        return Vector3D(v1.x - v2.x, v1.y - v2.y, v1.z - v2.z)
    
    @staticmethod
    def dot_product(v1, v2):
        return super().dot_product(v1, v2) + v1.z * v2.z
    
    @classmethod
    def distance(cls, v1, v2):
        radicando = substract(v1, v2)
        return radicando.module()
```

```python
class Pawns():
    
    def __init__(self):
        self.letters = []
    
    
    def addPawn(self, c):
        """
        A√±ade la ficha c a la lista de fichas letters
        Args:
            c: str (ficha)
        """
        self.letters.append(c)
    
    def addPawns(self, c, n):
        """
        A√±ade la ficha c a la lista de fichas letters n veces
        Args:
            c: str (ficha)
            n: int
        """
        for i in range(n):
            self.addPawn(c)
    
    def createBag(self):
        """
        Crea el saco de fichas
        """
        import pandas as pd
        datos_fichas = pd.read_csv('/content/drive/MyDrive/CursoPythonDeLaA-Z/csv/bag_of_pawns.csv')
    
        for ficha in datos_fichas.itertuples():
            self.addPawns(ficha[1], ficha[2]) 
    
    
    def showPawns(self):
        """
        Muestra por pantalla las fichas de Pawn
        """
        fichas_freq = self.getFrequency()
        fichas_freq.showFrequency()
                
        
    def takeRandomPawn(self):
        """
        Toma una ficha del saco de fichas y la elimina de la lista letters
        """
        import random
        ficha = self.letters.pop(random.randint(0, len(self.letters) - 1))
        return ficha
            
    def getFrequency(self):
        """
        Obtiene la frecuencia de las fichas de Pawn y lo devuelve en formato clase TableFrequencie
        """
        freq = FrequencyTable()
        for l in self.letters:
            freq.update(l)
        return freq
        
    def takePawn(self, c):
        self.letters.remove(c)
        
    def getTotalPawns(self):
        """
        Obtenemos el total de fichas del objeto
        """
        return len(self.letters)    
        

class Word():
    
    def __init__(self):
        self.word = []
    
    def __str__(self):
        """
        Imprime la palabra en formato string
        """
        palabra = ''
        for l in self.word:
            palabra += l
        return palabra
    
    
    def areEqual(self, w):
        """
        Chequea si la palabra por parametro y la que invoca son iguales
        """
        return w.__str__() == self.__str__()
    
    def isEmpty(self):
        """
        Chequea si la palabra es vacia
        """
        return self.word == []
    
    @classmethod
    def readWord(cls):
        """
        Lee una palabra por teclado y la devuelve como un objeto Word
        """
        input_word = input().strip()
        w = Word()
        for c in input_word.upper():
          w.word.append(c)
        return w
        
    @staticmethod
    def readWordFromLine(fichero):
        """
        Lee una palabra del fichero y la devuelve como un objeto de Word
        """
        palabra = Word()
        line = fichero.readline()
        for letra in line[:-1]:
            palabra.word.append(letra)
        return palabra
        
    def getFrequency(self):
        freq = FrequencyTable()
        for l in self.word:
            freq.update(l)
        return freq
        
    def getLengthWord():
        return len(self.word)
        
class Dictionary():
    
    file_path = '/content/drive/MyDrive/CursoPythonDeLaA-Z/modules/archivostxt/dictionary.txt'
    
    @staticmethod
    def validateWord(word):
        """
        Chequea si word pertenece al txt dictionary devolviendo True or False
        """
        with open(Dictionary.file_path, 'r') as dicc:
            w = Word.readWordFromLine(dicc)
            while not (w.isEmpty() or word.areEqual(w)):
                w = Word.readWordFromLine(dicc)
        if w.isEmpty() and not word.areEqual(w):
            return False
        else:
            return True
        
class FrequencyTable():
    
    def __init__(self):
        self.letters = [chr(x) for x in range(ord("A"), ord("Z") + 1)]
        self.frequencies = [0]*len(self.letters)

    def showFrequency(self):
        for i in range(len(self.letters)):
            if self.frequencies[i] != 0:
                print('La frecuencia de la letra {} es {}'.format(self.letters[i], self.frequencies[i]))

    @staticmethod
    def isSubset(fq1, fq2):
        es_subconjunto = True
        for i in range(len(fq1.frequencies)):
            if fq1.frequencies[i] > fq2.frequencies[i]:
                es_subconjunto = False
                break
        return es_subconjunto

    def update(self, c):
        indice = self.letters.index(c)
        self.frequencies[indice] += 1
        
class Board():
    
    def __init__(self):
        self.board = [[" " for j in range(15)] for i in range(15)]
        self.totalWords = 0
        self.totalPawns = 0
    
    def showBoard(self):
        #print(self.board)
        print("\n ", end = " ")
        for n in range(len(self.board)):
          print("{}{} ".format(0 if n <= 9 else "", n), end = " ")
        print("\n+" + "---+" * len(self.board))
        for i in range(len(self.board)):
          print("|", end = " ")
          for j in range(len(self.board)):
            print(self.board[i][j] + " |", end = " ")
          print("{}{}".format(0 if i <= 9 else "", i), end = " ")
          print("\n+" + "---+" * len(self.board))
          
    
    def placeWord(self, player_pawns, word, x, y, direction):
        for letter in word.word:
            if letter != self.board[x][y]:
              player_pawns.takePawn(letter)
              self.totalPawns += 1
              self.board[x][y] = letter
                
            if direction == "V":
              x += 1
            if direction == "H":
              y += 1
                
        self.totalWords += 1   
        
    def isPossible(self, word, x, y, direction):
        message = ''
        x1 = x
        y1 = y
        
        #caso primer turno
        if self.totalWords == 0:
            message = 'Ninguna ficha pasa por la casilla central'
            if direction == 'V':
                if y != 7:
                    return (False, message)
                elif x + word.getLengthWord(word) - 1 < 7 or x > 7:
                    return (False, message)
            
            if direction == 'H':
                if x != 7:
                    return (False, message)
                elif y + word.getLengthWord(word) - 1 < 7 or y > 7:
                    return (False, message)
        
        #caso limites del tablero
        message = 'La palabra sobrepasa los limites del tablero'
        if direction == 'H' and x + word.getLengthWord(word) - 1 > 14:
            return (False, message)
        if direction == 'V' and y + word.getLengthWord(word) - 1 > 14:
            return (False, message)
        if x < 0 or y < 0 or x > 14 or y > 14:
            return (False, message)
            
        #caso se debe usar una ficha ya existente en el tablero
        if self.totalWords > 0:
            lugares = []
            for c in word.word:
                if self.board[x][y] == " ":
                    lugares.append(c)
                  
                if direction == "V":
                    x += 1
                if direction == "H":
                    y += 1
            
            if len(lugares) == word.getLengthWord():
                message = "No se est√° utilizando ninguna ficha del tablero"
                return (False, message)

        #caso casillas libres
        x = x1
        y = y1
        for l in word.word:
            if self.board[x][y] != " " and self.board[x][y] != c:
                message = "Hay una ficha diferente ocupando una posici√≥n"
                return (False, message)
            if direction == "V":
                x += 1
            if direction == "H":
                y += 1
        
        #caso una ficha por casilla
        x = x1
        y = y1
        palabra = []
        for l in word.word:
            if self.board[x][y] == l:
                palabra.append(l)
            if direction == "V":
                x += 1
            if direction == "H":
                y += 1
        
        if len(palabra) == word.getLengthWord():
            message = "Recuerde que debe usar una ficha nueva en el tablero"
            return (False, message)
          
        #caso fichas adicionales
        message = "Hay fichas adicionales a principio o final de palabra"
        x = x1
        y = y1
        if direction == "V" and ((x != 0 and self.board[x - 1][y] != " ") or (x + word.getLengthWord() != 14 and self.board[x + word.getLengthWord()][y] != " ")):
            return (False, message)
        if direction == "H" and ((y != 0 and self.board[x][y - 1] != " ") or (y + word.getLengthWord() != 14 and self.board[x][y + word.getLengthWord()] != " ")):
            return (False, message)
      
        message = "La palabra se puede agregar al tablero"
        return (True, message)
      
    def getPawns(self, word, x, y, direction):
        
        letrasNec = Word()
        esPosible, message = self.isPossible(word, x, y, direction)
        
        if not esPosible:
            print(message)
        else:
            for c in word.word:
                if self.board[x][y] != c:
                    letrasNec.word.append(c)
                if direction == "V":
                    x += 1
                if direction == "H":
                    y += 1
                    
        return letrasNec
```
